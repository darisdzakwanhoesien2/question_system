# Exam Analysis Notebook

This notebook provides tools for analyzing exam performance data, generating insights, and visualizing results.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
from pathlib import Path
import numpy as np
from datetime import datetime

# Set up plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
```

## Load Exam Results Data

```python
def load_exam_results(results_dir: Path):
    """Load all exam results from storage directory."""
    results = []
    for file_path in results_dir.glob('*.json'):
        with open(file_path, 'r') as f:
            result = json.load(f)
            results.append(result)
    return results

# Load results
results_dir = Path('../storage/results')
exam_results = load_exam_results(results_dir)
print(f"Loaded {len(exam_results)} exam results")
```

## Performance Analysis

```python
def analyze_performance(results):
    """Analyze overall performance metrics."""
    if not results:
        return {}
    
    scores = [r.get('percentage_score', 0) for r in results]
    
    return {
        'average_score': np.mean(scores),
        'median_score': np.median(scores),
        'highest_score': max(scores),
        'lowest_score': min(scores),
        'total_exams': len(results)
    }

performance_stats = analyze_performance(exam_results)
performance_stats
```
